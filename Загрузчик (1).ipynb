```python
import numpy as np
import pandas as pd
import yfinance as yf
import os
```

```python
total_path = os.getcwd()

curr_path = total_path
data_path_m = curr_path + '//'+'Data//minutes'
data_path_h = curr_path + '//'+'Data//hours'
data_path_d = curr_path + '//'+'Data//days'
data_path_ds = curr_path + '//'+'Data//datasets'
```

```python
def download_ticker_full(ticker, period, start_date, end_date):
    
    # кастомная функция загрузки данных по тикеру.
    # принимает: название тикера
    # возвращает: датасет минутных данных по тикеру. Артефакт, раньше функция была сложнее (выносить код обратно не стал). 
    
    
    # загрузка минутных данных
    if period == 'm':
        res = yf.download(ticker, interval = "1m",start= start_date, end= end_date)
        res = res.reset_index()
        
    elif period == 'h':
        res = yf.download(ticker, interval = "1h",start= start_date, end= end_date)
        res = res.reset_index()
        
    elif period == 'd':
        res = yf.download(ticker, interval = "1d",start= start_date, end= end_date)
        res = res.reset_index()
    
    return res
```

```python
ticker_list = pd.read_excel(curr_path+'//cripto_list.xlsx') # грузим список тикеров
ticker_list = list(ticker_list['Тикер'])
ticker_list = [i+'-USD' for i in ticker_list]
```

```python
empty_tikers = []

periods = [["2022-10-22", "2022-10-27"],["2022-10-27", "2022-11-03"],["2022-11-03", "2022-11-10"], ["2022-11-10", "2022-11-17"]]
```

```python
os.chdir(data_path_m)  # грузим данные
for i in ticker_list:
    print(i)
    ticker_df = pd.DataFrame()
    for j in periods:
        df = download_ticker_full(ticker_list[0],'m', j[0],j[1])
        ticker_df = ticker_df.append(df)

    ticker_df = ticker_df.reset_index(drop = True)
    if len(df) > 0:
         df.to_csv(i[:-4]+'.csv', index = False)
os.chdir(total_path) 
```
